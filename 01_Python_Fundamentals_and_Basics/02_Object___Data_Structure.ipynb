- Strings are the sequence of characters using the syntax of either single quotes or double.


```
"hello World"
'hello'
```

- A string is a **sequence** of charactes.

- It is **immutable**.

- It is created by enclosing text with single, double or triple quotes

````
word = "Python"

print(word.lower())
print(word.upper())
print(len(word))
print(word.startswith("P"))
print(word.endswith("n"))

# Slicing the string
print(word[0:5])

# Coverting string to a list of characters
print(list(word))
````

#### <span style="display: inline-block; color: #fff; background: linear-gradient(135deg, #1e3a5f, #1565c0); padding: 7px 16px; border-radius: 8px; font-size: 16px; font-weight: 700; letter-spacing: 0.3px;"> .strip() — Remove Leading/Trailing Spaces </span>

````
this_string = "    python is easy          "
print(this_string)
print(this_string.strip())
````

#### <span style="display: inline-block; color: #fff; background: linear-gradient(135deg, #1e3a5f, #1565c0); padding: 7px 16px; border-radius: 8px; font-size: 16px; font-weight: 700; letter-spacing: 0.3px;"> .replace() — Replace Substrings </span>

````
this_str = "python is easy"
new_str1 = this_str.replace("easy", "") # this wiil replace easy with empty space
new_str2 = this_str.replace("easy", "hard") # this wiil replace easy with hard
print(new_str1)
print(new_str2)
````

#### <span style="display: inline-block; color: #fff; background: linear-gradient(135deg, #1e3a5f, #1565c0); padding: 7px 16px; border-radius: 8px; font-size: 16px; font-weight: 700; letter-spacing: 0.3px;"> .split() — Split into List </span>

````
# splitting the text
sentence = "A string is a sequence of charactes"
words = sentence.split(" ")

print(words)
print(len(words))
````

#### <span style="display: inline-block; color: #fff; background: linear-gradient(135deg, #1e3a5f, #1565c0); padding: 7px 16px; border-radius: 8px; font-size: 16px; font-weight: 700; letter-spacing: 0.3px;"> Advanced String Methods </span>


````
# Counting and Finding
s = "Hello World"
print(f"Count of 'l': {s.count('l')}")
print(f"Index of 'World': {s.find('World')}")

# Join and Split
words = ['Python', 'is', 'awesome']
sentence = " ".join(words)
print(f"Joined: {sentence}")
print(f"Split back: {sentence.split(' ')}")

# Numeric Checks
print(f"'123'.isdigit(): {'123'.isdigit()}")
print(f"'abc'.isalpha(): {'abc'.isalpha()}")
````

${\color{red}{\textbf{Note:}}}$ `.split` functon convert the string into `list` so the opearion like `.upper`, `.lower`, `.startswith` or `.endswith` won't work.

The output shows the type of a variable as a string (`str`). To fix this We need to convert user input explicitly to `integer` and `float` to check if it’s a `number`. If the input string is a `number`, It will get converted to `int` or `float` without exception.

````
def check_user_input(input):
    try:
        # Convert it into integer
        val = int(input)
        print("Input is an integer number. Number = {}".format(val))
    except ValueError:
        try:
            # Convert it into float
            val = float(input)
            print("Input is a float  number. Number = {}".format(val))
        except ValueError:
            print("No.. input is not a number. It's a string")


input1 = input("Enter your int ")
check_user_input(input1)

input2 = input("Enter any float ")
check_user_input(input2)

input2 = input("Enter the last string ")
check_user_input(input2)
````

  📊 7.1. Arithmetic Operators
</span>

> Used to perform common mathematical operations: `+`, `-`, `*`, `/`, `//`, `%`, `**`

````
a, b = 10, 3

print(f'Addition       : {a} + {b}  = {a + b}')
print(f'Subtraction    : {a} - {b}  = {a - b}')
print(f'Multiplication : {a} * {b}  = {a * b}')
print(f'Division       : {a} / {b}  = {a / b:.4f}')
print(f'Floor Division : {a} // {b} = {a // b}')
print(f'Modulus        : {a} % {b}  = {a % b}')
print(f'Exponentiation : {a} ** {b} = {a ** b}')
````

> Used to assign values to variables: `=`, `+=`, `-=`, `*=`, `/=`, `//=`, `%=`, `**=`

````
x = 10
print(f'Initial x = {x}')

x += 5;  print(f'x += 5  → {x}')
x -= 3;  print(f'x -= 3  → {x}')
x *= 2;  print(f'x *= 2  → {x}')
x //= 4; print(f'x //= 4 → {x}')
x **= 3; print(f'x **= 3 → {x}')
x %= 5;  print(f'x %= 5  → {x}')
````

  📊 7.3. Comparison Operators
</span>

> Used to compare two values and always return a **Boolean** (`True` / `False`): `==`, `!=`, `>`, `<`, `>=`, `<=`

````
a, b = 10, 20

print(f'{a} == {b}  → {a == b}')   # Equal
print(f'{a} != {b}  → {a != b}')   # Not Equal
print(f'{a} >  {b}  → {a > b}')    # Greater Than
print(f'{a} <  {b}  → {a < b}')    # Less Than
print(f'{a} >= {b}  → {a >= b}')   # Greater or Equal
print(f'{a} <= {b}  → {a <= b}')   # Less or Equal
````

  📊 7.5. Identity Operators
</span>

> Test if two variables point to the **same object in memory**: `is`, `is not`

````
a = [1, 2, 3]
b = a          # b points to the same object as a
c = [1, 2, 3]  # c is a NEW list with same values

print(f'a is b     → {a is b}')      # True  - same object
print(f'a is c     → {a is c}')      # False - different objects
print(f'a == c     → {a == c}')      # True  - same values
print(f'a is not c → {a is not c}')  # True
````

  📊 7.7. Bitwise Operators
</span>

> Operate on bits directly: `&` (AND), `|` (OR), `^` (XOR), `~` (NOT), `<<` (left shift), `>>` (right shift)

````
a, b = 0b1010, 0b1100  # 10 and 12 in binary

print(f'a       = {bin(a)} ({a})')
print(f'b       = {bin(b)} ({b})')
print(f'a & b   = {bin(a & b)} ({a & b})  # AND')
print(f'a | b   = {bin(a | b)} ({a | b})  # OR')
print(f'a ^ b   = {bin(a ^ b)} ({a ^ b})   # XOR')
print(f'~a      = {bin(~a)} ({~a})   # NOT')
print(f'a << 1  = {bin(a << 1)} ({a << 1})  # Left Shift')
print(f'a >> 1  = {bin(a >> 1)} ({a >> 1})    # Right Shift')
````

#### <span style="display: inline-block; color: #fff; background: linear-gradient(135deg, #1e3a5f, #1565c0); padding: 7px 16px; border-radius: 8px; font-size: 16px; font-weight: 700; letter-spacing: 0.3px;"> Operator Precedence (PEMDAS/BODMAS) </span>

| Priority | Operator | Description |
| :---: | :--- | :--- |
| 1 | `**` | Exponentiation |
| 2 | `~`, `+`, `-` | Unary ops |
| 3 | `*`, `/`, `//`, `%` | Multiply/Divide |
| 4 | `+`, `-` | Add/Subtract |
| 5 | `>>`, `<<` | Bitwise Shift |
| 6 | `&` | Bitwise AND |
| 7 | `^`, `\|` | Bitwise XOR/OR |
| 8 | `<=`, `>=`, `<`, `>` | Comparison |
| 9 | `==`, `!=` | Equality |
| 10 | `=`, `+=`, ... | Assignment |

````
# Operator precedence examples
print(2 + 3 * 4)    # 14 -- * before +
print((2 + 3) * 4)  # 20 -- () first
print(2 ** 3 ** 2)  # 512 -- ** is right-associative: 2 ** (3**2)
print(10 % 3 + 1)   # 2  -- % before +
````

#### <span style="display: inline-block; color: #fff; background: linear-gradient(135deg, #1e3a5f, #1565c0); padding: 7px 16px; border-radius: 8px; font-size: 16px; font-weight: 700; letter-spacing: 0.3px;"> Continue Statement </span>

````
i = 0
while i < 6:
    i += 1
    if i == 4:   # we can stop the current iteration
        continue # continue with the next:
    print(i)   
````

````
i = 1
while i < 6:
    print(i)
    i += 1
else:
    print('i is no longer less then 6')
````

A **for loop** is used for iterating over a sequence (that is either a `list`, a `tuple`, a `dict`, a `set`, or a `str`).

````
# Looping over a list

thelist = ['apple', 'banana', 'orange']

for i in thelist:
    print(i)
````

````
# Looping over a string

for i in 'cat':
    print(i)
````

````
# The break Statement

thelist = [1, 2, 3, 4, 5]
for i in thelist:
    print(i)
    if i == 4:
        break
````

````
thelist = [1, 2, 'a', 4, 5, 'b']

for i in thelist:
    if i == 'a':
        break
    print(i) 
````

````
# Continue Statement

thelist = [1, 2, 'a', 4, 5, 'b']

for i in thelist:
    if i == 'a':
        continue
    print(i)
````
