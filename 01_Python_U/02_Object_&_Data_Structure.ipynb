#### a. `while ` __Loop__

**The break Statement**

````
i = 1
while i < 6:
    i += 1
    print(i)
    if i == 4:
        break
    
````

````
i = 1
while i < 6:
    print(i)
    if i == 4:
        break
    i += 1
````

**The continue Statement**

````
i = 0
while i < 6:
    i += 1
    if i == 4:   # we can stop the current iteration
        continue # continue with the next:
    print(i)
    
      
````

**The else Statement**

````
i = 1
while i < 6:
    print(i)
    i += 1
else:
    print('i is no longer less then 1')
````

#### b. `for` __loop__

A **for loop** is used for iterating over a sequence (that is either a **list, a tuple, a dictionary, a set, or a string**).

````
# itration over a list
thelist = ['apple', 'banana', 'orange']
for x in thelist:
    print(x)
````

````
# Looping Through a String
for x in 'cat':
    print(x)
````

````
# The break Statement
thelist = [1, 2, 3, 4, 5]
for x in thelist:
    print(x)
    if x == 4:
        break
````

````
thelist = [1, 2, 3, 4, 5]
for x in thelist:
    if x == 4:
        break
    print(x) 
````

````
# Continue Statement
thelist = [1, 2, 'a', 4, 5, 'b']
for x in thelist:
    if x == 'a':
        continue
    print(x)
    
````

````
# Range function
for x in range(0,10,4):
    print(x)
else:
    print('Finally Finished')
````

````
# Nested loops
lis1 = [1, 2, 3]
lis2 = ['a', 'b', 'c']

for x in lis1:
    for y in lis2:
        print(x, y)
````

**for loops** cannot be **empty**, but if you for some reason you want have a for loop with **no content** for later use, put in the pass statement to avoid getting an error.

````
# Pass Statement
for x in [0, 1, 2]:
  pass
````

`shuffle()`:
This function __shuffle__ the list and randomly arrange them

````
list_1 = [1, 3, 5, 10, 4] 

print ("list before shuffling : ", end="") 

for x in range(0, len(list_1)): 
   print (list_1[x], end=" ") 
print("\n") 
random.shuffle(list_1) 
print ("list after shuffling  : ", end="") 
for x in range(0, len(list_1)): 
   print (list_1[x], end=" ")
````
