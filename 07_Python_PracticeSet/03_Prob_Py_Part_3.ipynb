  Que:5 Write a program to find the sum of first n natural numbers using while loop.
</span>

````python
n = int(input("Enter the Number: "))

i = 1
sum = 0 #Note: for addition, we initialize the sum with zero.

while(i<=n):
    sum += i
    i += 1

print(sum)
````

**Arithmetic mean**

````python
n = int(input("Enter the Number: "))

i = 1
sum = 0 # Note: for addition, we initialize the sum with zero.

while(i <= n):
    sum += i
    i += 1

arithmetic_mean = sum / n
print(f"Arithmetic Mean is {sum}/{n} =", arithmetic_mean)
````

  Que:6 Write a program to calculate the factorial of a given number using for loop.

````python
# 5! = 1 X 2 X 3 X 4 X 5

n = int(input("Enter the Number: "))

product = 1 # Note: for product, we initialize the sum with one.

for i in range(1, n+1):
    product = product * i

print(product)
````

</span>

- for `n = 4`

```
   *        space=3, star=1
  ***       space=2, star=3
 *****      space=1, star=5
*******     space=0, star=7
 *****      space=1, star=5
  ***       space=2, star=3
   *        space=3, star=1

```


````python
n = int(input("Enter the Number: "))

# Upper part of the pattern
for i in range(1, n+1):
    print(" "*(n-i), end="")
    print("*"*(2*i-1), end="")
    print("")

# Lower part of the pattern
for j in range(n-1, 0, -1):
    print(" "*(n-j), end="")
    print("*"*(2*j-1), end="")
    print("")
````

</span>

```
***
* *
***
```

for `n = 3`.

```
****
*  *
*  *
****
```

for `n = 4`.

````python
n = int(input("Enter the Number: "))

# Upper part of the pattern
for i in range(1, n+1):
    if(i==1 or i==n):
        print("*"*n, end="")
    else:
        print("*", end="")
        print(" "*(n-2), end="")
        print("*", end="")
    print("")
````
