  🧾 Q01 — What are Decorators and how do they work?
</span>

A decorator is a function that takes another function and extends its behavior without explicitly modifying it. In Python, functions are first-class objects, meaning they can be passed as arguments and returned from other functions. Decorators use this to wrap a function, executing code before and after the wrapped function runs.

```python
def my_decorator(func):
    def wrapper(*args, **kwargs):
        print("Something is happening before the function is called.")
        result = func(*args, **kwargs)
        print("Something is happening after the function is called.")
        return result
    return wrapper

@my_decorator
def say_whee():
    print("Whee!")

say_whee()
```

  🧾 Q02 — Explain Generators and the `yield` keyword.
</span>

Generators are functions that return an iterable set of items, one at a time, in a special way. Instead of using `return` to return a value and exit, they use `yield`. When `yield` is called, the function's state is paused and saved, allowing it to resume exactly where it left off on the next iteration. This makes them incredibly memory-efficient for large datasets since they generate items on the fly instead of storing them all in memory.

```python
def fibonacci_generator(limit):
    a, b = 0, 1
    while a < limit:
        yield a
        a, b = b, a + b

# Using the generator
for num in fibonacci_generator(50):
    print(num, end=' ')
```

  🧾 Q03 — How do Context Managers (`with` statement) work?
</span>

Context managers allow you to allocate and release resources precisely when you want to. The most widely used example is the `with` statement when opening files. It ensures that the resource is properly closed/released even if an exception occurs. Under the hood, a context manager implements the `__enter__` and `__exit__` magic methods.

```python
class ManagedFile:
    def __init__(self, filename):
        self.filename = filename

    def __enter__(self):
        print(f"Opening file: {self.filename}")
        self.file = open(self.filename, 'w')
        return self.file

    def __exit__(self, exc_type, exc_val, exc_tb):
        print(f"Closing file: {self.filename}")
        if self.file:
            self.file.close()

# Testing the context manager
with ManagedFile('test.txt') as f:
    f.write('Hello, World!')
```

  🧾 Q04 — What are Metaclasses in Python?
</span>

In Python, a metaclass is the class of a class. Just as a class defines how an instance of the class behaves, a metaclass defines how a class behaves. A class is an instance of a metaclass (by default, the `type` metaclass). Metaclasses are used to intercept the creation of classes and modify them, which is useful for things like automatic registration, singleton patterns, or API validation.

```python
class SingletonMeta(type):
    _instances = {}
    def __call__(cls, *args, **kwargs):
        if cls not in cls._instances:
            cls._instances[cls] = super().__call__(*args, **kwargs)
        return cls._instances[cls]

class Singleton(metaclass=SingletonMeta):
    def __init__(self):
        print("Creating instance")

obj1 = Singleton()
obj2 = Singleton()
print("obj1 is obj2:", obj1 is obj2)  # Output: True
```

  🧾 Q05 — Explain Method Resolution Order (MRO) and `super()`.
</span>

Method Resolution Order (MRO) is the order in which Python looks for a method in a hierarchy of classes, especially in multiple inheritance. Python uses the C3 Linearization algorithm to build the MRO. The `super()` function allows you to call methods of the superclass in the MRO, ensuring that each parent's method is called exactly once in a diamond inheritance pattern.

```python
class A:
    def process(self):
        print("A's process()")

class B(A):
    def process(self):
        print("B's process()")
        super().process()

class C(A):
    def process(self):
        print("C's process()")
        super().process()

class D(B, C):
    def process(self):
        print("D's process()")
        super().process()

obj = D()
obj.process()
print("\nMRO for D:", [cls.__name__ for cls in D.mro()])
```

  🧾 Q06 — What is a Closure and what is the Late Binding issue?
</span>

A closure occurs when a nested function captures and remembers the local variables of its enclosing function, even after the enclosing function has finished executing. However, Python's closures exhibit "late binding", meaning the values of variables used in closures are looked up at the time the inner function is called, not when it is defined. This often causes unexpected behavior in loops.

```python
def create_multipliers():
    # Late binding issue: 'i' is looked up when the lambda is called, so all lambdas use the final value of 'i' (4)
    return [lambda x: i * x for i in range(5)]

def create_multipliers_fixed():
    # Fixed by binding 'i' as a default argument at definition time
    return [lambda x, i=i: i * x for i in range(5)]

broken = [m(2) for m in create_multipliers()]
fixed = [m(2) for m in create_multipliers_fixed()]

print("Broken (late binding):", broken)
print("Fixed:", fixed)
```

  🧾 Q07 — What is `__slots__` and why use it?
</span>

By default, Python stores instance attributes in a dynamic dictionary (`__dict__`). This allows adding attributes at runtime but uses a lot of memory. Defining `__slots__` in a class tells Python to allocate space for a fixed set of attributes, preventing the creation of `__dict__`. This significantly reduces memory usage and speeds up attribute access for classes with millions of instances.

```python
import sys

class WithoutSlots:
    def __init__(self, name, identifier):
        self.name = name
        self.identifier = identifier

class WithSlots:
    __slots__ = ['name', 'identifier']
    def __init__(self, name, identifier):
        self.name = name
        self.identifier = identifier

obj1 = WithoutSlots("Test", 123)
obj2 = WithSlots("Test", 123)

# We check memory size of instances
print("Memory without slots:", sys.getsizeof(obj1) + sys.getsizeof(obj1.__dict__))
print("Memory with slots:", sys.getsizeof(obj2))  # No __dict__ to add
```

  🧾 Q08 — How does Asynchronous Programming (`asyncio`) work in Python?
</span>

Asynchronous programming allows a program to handle multiple I/O-bound tasks concurrently without using multiple threads. Python's `asyncio` uses an event loop to manage execution. When a coroutine reaches an `await` expression, it yields control back to the event loop, which can run other tasks while waiting for the I/O operation (like a network request or file read) to finish.

```python
import asyncio
import time

async def fetch_data(id, delay):
    print(f"Task {id}: Starting fetch...")
    await asyncio.sleep(delay)  # Simulates non-blocking I/O (e.g., network request)
    print(f"Task {id}: Fetch completed.")
    return f"Data {id}"

async def main():
    start = time.time()
    # Run tasks concurrently
    results = await asyncio.gather(
        fetch_data(1, 2),
        fetch_data(2, 3),
        fetch_data(3, 1)
    )
    print("Results:", results)
    print(f"Total time: {time.time() - start:.2f} seconds (Note: Much less than 2+3+1=6s)")

# To run in Jupyter (which already has an event loop), we use await directly:
# await main()
# (In a script, you would use asyncio.run(main()))
```

```python
import math

# Original behavior
print("Original pi:", math.pi)

# Monkey patching a built-in module attribute
math.pi = 3.0

def area_of_circle(radius):
    return math.pi * (radius ** 2)

print("Monkey patched area (r=10):", area_of_circle(10)) # Uses 3.0 instead of 3.14159...

# Restore original to prevent issues elsewhere
math.pi = 3.141592653589793
```

  🧾 Q10 — Explain the Descriptor Protocol (`__get__`, `__set__`, `__delete__`).
</span>

Descriptors are classes that implement the descriptor protocol: `__get__`, `__set__`, or `__delete__`. They allow you to customize how attributes are accessed or mutated on an object. Python uses descriptors heavily under the hood—features like properties (`@property`), class methods, and static methods are all implemented using descriptors.

```python
class ValidatedNumber:
    def __init__(self, min_val, max_val):
        self.min_val = min_val
        self.max_val = max_val
        self.name = None

    def __set_name__(self, owner, name):
        self.name = name

    def __get__(self, instance, owner):
        if instance is None: return self
        return instance.__dict__.get(self.name)

    def __set__(self, instance, value):
        if not (self.min_val <= value <= self.max_val):
            raise ValueError(f"{self.name} must be between {self.min_val} and {self.max_val}")
        instance.__dict__[self.name] = value

class Person:
    age = ValidatedNumber(0, 120)

p = Person()
p.age = 30
print("Age is:", p.age)

try:
    p.age = 150 # Triggers __set__ logic
except ValueError as e:
    print("Error:", e)
```

  🧾 Q25 — Explain `__iter__` and `__next__`.
</span>

These two methods implement the iterator protocol. `__iter__` returns the iterator object itself, and `__next__` returns the next value or raises `StopIteration`.

```python
class Counter:
    def __init__(self, limit):
        self.limit, self.val = limit, 0
    def __iter__(self):
        return self
    def __next__(self):
        if self.val >= self.limit: raise StopIteration
        self.val += 1
        return self.val
```

  🧾 Q100 — What is `typing.Protocol` and how does it relate to Structural Subtyping?
</span>

Introduced in Python 3.8, `Protocol` allows for static duck typing (structural subtyping). Instead of forcing a class to explicitly inherit from a base class to satisfy a type checker, you define a `Protocol`. If a class implements the methods defined in the `Protocol`, the type checker considers it a valid match.

```python
from typing import Protocol

class SupportsQuack(Protocol):
    def quack(self) -> str: ...

class Duck:
    def quack(self) -> str: return 'Quack!'

def make_it_quack(obj: SupportsQuack):
    print(obj.quack())

# Duck doesn't inherit from SupportsQuack, but satisfies the Protocol
d = Duck()
make_it_quack(d)
```
