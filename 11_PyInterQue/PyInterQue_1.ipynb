  🧾 21. Create generator to produce first n prime numbers
</span>

```python
def isprime(num):
    for i in range(2, num):
        if num % i == 0:
            return False
    return True

def prime_generator(n):
    num = 2
    while n:
        if isprime(num):
            yield num
            n -= 1
        num += 1

x = int(input("Enter no of prime numbers required: "))
it = prime_generator(x)
for e in it:
    print(e, end=" ")
```

  🧾 32. What is python decorator ?
</span>

```python
# Function decorator
def welcome(fx):
    def mfx(*t, **d):
        print("Before hello function")
        fx(*t, **d)
        print("Thanks for using the function")
    return mfx

@welcome
def hello():
    print("Hello")

@welcome
def add(a, b):
    print(a + b)

hello()
add(1, 3)

# Class Decorator
class Calculator:
    def __init__(self, func):
        self.function = func
    def __call__(self, *t, **d):
        result = self.function(*t, **d)
        return result ** 2

@Calculator
def add(a, b):
    return a + b

print(add(10, 20))
```

  🧾 34. What are generators in python? Create a generator for first n natural even numbers.

```python
def evenNum(n):
    i = 1
    while n:
        yield 2 * i
        i += 1
        n -= 1

it = evenNum(10)
even_list = []
while True:
    try:
        even_list.append(next(it))
    except StopIteration:
        break
print(even_list)
```

  🧾 86. Operator overloading in python:

```python
class Point:
    def __init__(self, x=0, y=0):
        self.x = x
        self.y = y
    def __add__(self, other):
        return Point(self.x + other.x, self.y + other.y)
    def __mul__(self, other):
        if isinstance(other, Point):
            return Point(self.x * other.x, self.y * other.y)
        else:
            return Point(self.x * other, self.y * other)
    def __rmul__(self, other):
        return Point(self.x * other, self.y * other)
    def __repr__(self):
        return "{0}, {1}".format(self.x, self.y)

p1 = Point(1, 2)
p2 = Point(3, 4)
print(p1 + p2)
print(p1 * p2)
```

```python
# Class-based Context Manager
class FileOpenContext:
    def __init__(self, filename, mode):
        self.filename = filename
        self.mode = mode
        self.file = None
    def __enter__(self):
        self.file = open(self.filename, self.mode)
        return self.file
    def __exit__(self, exc_type, exc_val, exc_tb):
        if self.file:
            self.file.close()

# Writing with context manager
with FileOpenContext("test_ctx.txt", "w") as f:
    f.write("Hello from context manager!")

# Generator-based Context Manager using contextlib
from contextlib import contextmanager

@contextmanager
def open_file(filename, mode):
    f = open(filename, mode)
    try:
        yield f
    finally:
        f.close()

with open_file("test_ctx.txt", "r") as f:
    print(f.read())
```

```python
nested_list = [[1, 2, 3], [4, 5], [6, 7, 8, 9]]

# 1. Using list comprehension
flat_list = [item for sublist in nested_list for item in sublist]
print("Flat List:", flat_list)

# 2. Using recursive function for deeply nested lists
def flatten_deep(lst):
    flat = []
    for item in lst:
        if isinstance(item, list):
            flat.extend(flatten_deep(item))
        else:
            flat.append(item)
    return flat

deeply_nested = [1, [2, [3, 4], 5], 6]
print("Deeply Nested Flat List:", flatten_deep(deeply_nested))
```
